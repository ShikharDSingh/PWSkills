Question 1: What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to draw conclusions about a population based on sample data.
  - It allows you to assess whether there is enough evidence in the sample data to support a specific hypothesis about the population or not.
  - In hypothesis testing We consider two hypotheses:-
    - Null Hypothesis:- Which states the claim you are trying to support.
    - Alternative Hypothesis:- Which contradicts the null hypothesis.
  - There are two types of errors in hypothesis testing:-
    - Type I Error:- Rejecting the null hypothesis when it is actually true.
    - Type II Error:- Failing to reject the null hypothesis when it is actually false.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - Null Hypothesis:-
    - The null hypothesis is the default assumption that we test against in hypothesis testing.
    - The null hypothesis is assumed to be true until its proven false.
    - The null hypothesis typically has a neutral position suggesting no change or difference from a baseline.
  - Alternative hypothesis:-
    - It contradicts the null hypothesis and we try to prove Alternative hypothesis.
    - The alternative hypothesis is what you want to prove in your test.
    - If there is enough evidence against the null hypothesis you would reject null hypothesis in favor of the alternative hypothesis.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
  - The significance level tells us how much evidence we need before we can confidently reject the null hypothesis.
  - It is denothed by alpha α.
  - The significance level is the probability of rejecting the null hypothesis when it is actually true.
  - It represents the acceptable risk of making a Type I error.
  - The significance level acts as point that decides whether the results are statistically significant or not.
  - If the p-value is smaller than or equal to α  you reject the null hypothesis.
  - If the p-value is greater than α you fail to reject the null hypothesis.

Question 4: What are Type I and Type II errors? Give examples of each.
  - Type I error and a Type II error refer to incorrect conclusions that can be drawn when testing the null hypothesis.
  - Type I error:-
    - A Type I error occurs when the null hypothesis is rejected when it is actually true.
    - The probability of making a Type I error is denoted by alpha α which is the level of significance of the test.
  - Type II error:-
    - A Type II error occurs when the null hypothesis is not rejected when it is actually false.
    - The probability of making a Type II error is denoted by beta β.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
  - Z-Test:-
    - A Z-test is used when the population standard deviation is known or the sample size is large (n ≥ 30).
    - The Z-test assumes that the standard deviation of the population is known.
    - It is used on sample data because collecting data from full popluation is not possible.
    - It assumes the data follows a normal distribution.
  - T-Test:-
    - A T-test is used when the population standard deviation is unknown, and it is typically used when the sample size is small (n less than 30).
    - The T-test relies on the t-distribution, which is a modified form of the normal distribution.
    - It is used when standard deviation of a populaton is unknown and population size is less than 30.

```python
#Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5
size = 1000

binomial_samples = np.random.binomial(n, p, size)

plt.figure(figsize=(8, 6))
plt.hist(binomial_samples, bins=np.arange(n+2)-0.5, edgecolor='black', alpha=0.7)
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.xticks(np.arange(n+1))
plt.show()
```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import scipy.stats as stats


sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
population_mean = 50
population_std_dev = 1

z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

reject_null = abs(z_statistic) > z_critical

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"Critical Z-value: {z_critical:.4f}")
print(f"Reject null hypothesis: {reject_null}")

if reject_null:
    print("We reject the null hypothesis. There is a significant difference between the sample mean and the population mean.")
else:
    print("We fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

mean = 50
std_dev = 5
n = 1000

simulated_data = np.random.normal(loc=mean, scale=std_dev, size=n)

sample_mean = np.mean(simulated_data)
sample_std_dev = np.std(simulated_data, ddof=1)
standard_error = sample_std_dev / np.sqrt(n)

z_critical = stats.norm.ppf(1 - 0.025)

margin_of_error = z_critical * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

plt.figure(figsize=(8, 6))
plt.hist(simulated_data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {sample_mean:.2f}')
plt.axvline(lower_bound, color='green', linestyle='dashed', linewidth=2, label=f'95% CI Lower Bound: {lower_bound:.2f}')
plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label=f'95% CI Upper Bound: {upper_bound:.2f}')
plt.title('Histogram of Simulated Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)

    z_scores = (data - mean) / std_dev

    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return z_scores

data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_z_scores(data)

print(f"Z-scores: {z_scores}")
```
