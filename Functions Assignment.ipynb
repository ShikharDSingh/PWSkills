```python
# Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum

list1 = [1,2,3,4,5]
print(sum_even_numbers(list1))
```

```python
# Create a Python function that accepts a string and returns the reverse of that string
def reverse_string(string):
    return string[::-1]

print(reverse_string("Hello"))
```

```python
# Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    squares = []
    for num in numbers:
        squares.append(num**2)
    return squares

list1 = [1,2,3,4,5]
print(square_numbers(list1))
```

```python
# Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True
print(is_prime(17))
print(is_prime(101))
```

```python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.i = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        self.i += 1
        current = self.a
        self.a, self.b = self.b, self.a + self.b
        return current

fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    power = 1
    for i in range(n):
        yield power
        power *= 2

for power in powers_of_two(5):
    print(power, end=" ")
```

```python
# Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
for line in read_file_lines('file.txt'):
    print(line)
```

```python
# Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
list1 = [(1,2), (3,4), (5,6)]
sorted_list = sorted(list1, key=lambda x: x[1])
print(sorted_list)
```

```python
# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
temps = [0, 10, 20, 30, 40]
def c_to_f(c):
    return (c * 9/5) + 32
fahrenheit_temps = list(map(c_to_f, temps))
print("Celsius:", temps)
print("Fahrenheit:", fahrenheit_temps)
```

```python
# Create a Python program that uses `filter()` to remove all the vowels from a given string.
str1 = "Hi I am Shikhar Singh"
vowels = ['a', 'e', 'i', 'o', 'u']
def remove_vowels(string):
    return ''.join(filter(lambda x: x.lower() not in vowels, string))
print(remove_vowels(str1))
```

```python
# Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
# product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
# order is smaller than 100,00 
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]

result = list(map(
    lambda order: (
        order[0],
        order[2] * order[3] + (10 if (order[2] * order[3]) < 10000 else 0)
    ),
    orders
))

print(result)
```
