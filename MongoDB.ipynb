1.What are the key differences between SQL and NoSQL databases?

10.What are the ACID properties in the context of MongoDB transactions?
  - ACID refers to Atomicity, Consistancy, Isolation, Durability.
  - These are core concept of a transaction in a database.
  - Atomicity:-
    - If a transaction is performed it will complete fully or it will not perform at all.
    - If transaction fails in between it will be rollbact to its previous state.
  - Consistancy:-
    - Database will always be in a consistant state after every transaction.
  - Isolation:-
    - Concurrent transactions do not interfere with each other.
  - Durability:-
    - Once a transaction is committed, the data is permanently saved.

13.What is the difference between a primary and a secondary node in a replica set?
  - Primary Node:-
    - Only one node in a replica set can be the primary at any time.
    - The primary node handles all write operations and serves them to the clients
    - Once the primary node accepts a write it replicates that data to the secondary nodes.
  - Secondary Node:-
    - Secondary nodes are exact copies of the primary node’s data.
    - Secondary nodes are mainly for read operations
    - Secondary nodes replicate data from the primary node and dont perform write itself.

20.What are the main differences between capped collections and regular collections?
  - Capped Collections:-
    - It have a fixed size.
    - It is ideal for high throughput logging and real time data feeds.
    - Cannot delete specific documents.
    - Updates must not increase document size.
  - Regular Collections:-
    - It have no fixed size limit.
    - Suitable for most standard use cases like user profiles, product catalogs, orders, etc.

    - Fully supports insert, update, and delete operations.

```python
# 7. Update all orders with Ship Mode as "First Class" to "Premium Class.
db = client['superstore_db']
collection = db['orders']

result = collection.update_many(
    {'Ship Mode': 'First Class'},
    {'$set': {'Ship Mode': 'Premium Class'}}
)

print(f"Updated {result.modified_count} documents.")
```

```python
# 8.Delete all orders where Sales is less than 50
db = client['superstore_db']
collection = db['orders']

result = collection.delete_many({'Sales': {'$lt': 50}})

print(f"Deleted {result.deleted_count} documents where Sales < 50.")
```

```python
# 9.Use aggregation to group orders by Region and calculate total sales per region

db = client['superstore_db']
collection = db['orders']

pipeline = [
    {
        '$group': {
            '_id': '$Region',
            'TotalSales': {'$sum': '$Sales'}
        }
    },
    {
        '$sort': {'TotalSales': -1}
    }
]

results = collection.aggregate(pipeline)

print("Total Sales by Region:")
for doc in results:
    print(f"Region: {doc['_id']}, Total Sales: {doc['TotalSales']:.2f}")
```

```python
# 10.Fetch all distinct values for Ship Mode from the collection
db = client['superstore_db']
collection = db['orders']

distinct_ship_modes = collection.distinct('Ship Mode')

print("Distinct Ship Modes:")
for mode in distinct_ship_modes:
    print(f"- {mode}")
```

```python
# 11.Count the number of orders for each category.

pipeline = [
    {
        '$group': {
            '_id': '$Category',
            'OrderCount': {'$sum': 1}
        }
    },
    {
        '$sort': {'OrderCount': -1}
    }
]

results = collection.aggregate(pipeline)

print("Number of Orders per Category:")
for doc in results:
    print(f"- {doc['_id']}: {doc['OrderCount']} orders")
```
