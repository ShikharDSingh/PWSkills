**Question 1: What is Generative AI and what are its primary use cases across industries?**

    - Finance Industry:-

      - It can creates advertisements, slogans, and marketing content.

      - It generates programming code automatically.
      - It helps developers with debugging and code suggestions.

**Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?**
  - Probabilistic modeling is a statistical approach that represents uncertainty in data using probability distributions.
  - It helps generative models understand how data is produced and how different variables are related.
  - It plays a key role in training generative models to create realistic new data samples.
  - By learning probability distributions, generative models can generate new data samples similar to the training data.
  - It allows the model to handle randomness and uncertainty in real-world data.
  - It enables models to simulate realistic data samples.
  - How do these models differ from discriminative models:-
    - Its main goal is to generate new data whereas discriminative models predict labels.
    - Generative models learns full data distribution whereas discriminative models learn decision boundary.
    - It can create new samples whereas discriminative models cannot generate new data.

**Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?**
  - Autoencoders:-
    - An Autoencoder is a type of neural network used to encode input data into a smaller representation and then reconstruct it back to its original form.
    - It produces deterministic latent representations.
    - Its mainly used for dimensionality reduction and feature learning.
    - Reconstruction focuses on reproducing the same input rather than generating new variations.
  - Variational Autoencoders:-
    - A Variational Autoencoders (VAE) is an advanced version of an autoencoder that uses probabilistic modeling to generate new data samples.
    - It uses probabilistic latent variables.
    - It allows generation of diverse and realistic text samples.
    - It ensures a continuous and well-organized latent space.

**Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?**
  - Neural Machine Translation is a deep learning approach used to automatically translate text from one language to another.
  - Early NMT models used encoder decoder architectures, which sometimes struggled with long sentences.
  - To solve this problem, the Attention Mechanism was introduced.
  - Attention allows the model to focus on the most relevant words in the input sentence while generating each word in the translation.
  - Without attention, the model compresses the entire sentence into a single fixed vector, which can lose information.
  - Attention allows the model to access all encoder outputs directly.
  - The model learns to focus on important words related to the current translation step.
  - Attention significantly improves accuracy and fluency of translations.
  - Helps the model understand relationships between words across the sentence.
  - Attention mechanisms later became the core idea behind advanced architectures like the Transformer Architecture.

```python
# Question 6: Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction:
'''["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]'''
# 1. Preprocess the data (tokenize and pad the sequences).
# 2. Build a basic VAE model for text reconstruction.
# 3. Train the model and show how it reconstructs or generates similar sentences.

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Lambda, RepeatVector, TimeDistributed
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K

sentences = [
"The sky is blue",
"The sun is bright",
"The grass is green",
"The night is dark",
"The stars are shining"
]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)
max_length = max(len(s) for s in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding="post")
vocab_size = len(tokenizer.word_index) + 1
print("Vocabulary Size:", vocab_size)
print("Padded Sequences:\n", padded_sequences)

X = padded_sequences
y = padded_sequences

latent_dim = 8
embedding_dim = 16
inputs = Input(shape=(max_length,))
x = Embedding(vocab_size, embedding_dim)(inputs)
x = LSTM(32)(x)
z_mean = Dense(latent_dim)(x)
z_log_var = Dense(latent_dim)(x)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon

z = Lambda(sampling)([z_mean, z_log_var])

decoder = RepeatVector(max_length)(z)
decoder = LSTM(32, return_sequences=True)(decoder)
outputs = TimeDistributed(Dense(vocab_size, activation="softmax"))(decoder)

vae = Model(inputs, outputs)

vae.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy"
)

vae.summary()
vae.fit(X, np.expand_dims(y, -1), epochs=200, verbose=1)

pred = vae.predict(X)
predicted_words = np.argmax(pred, axis=-1)
reverse_word_index = {v:k for k,v in tokenizer.word_index.items()}
for i in range(len(sentences)):
    decoded = []
    for idx in predicted_words[i]:
        word = reverse_word_index.get(idx, "")
        decoded.append(word)

    print("Original:", sentences[i])
    print("Reconstructed:", " ".join(decoded))
    print()
```

```python
# Question 7: Use a pre-trained GPT model (like GPT-2 or GPT-3) to translate a short English paragraph into French and German. Provide the original and translated text.

from transformers import pipeline

text = "Artificial Intelligence is transforming many industries by enabling automation."
generator = pipeline("text-generation", model="gpt2")

def translate_with_gpt2(input_text, target_lang):
    prompt = f"Translate English to {target_lang}:\nEnglish: {input_text}\n{target_lang}:"

    result = generator(prompt, max_new_tokens=40, num_return_sequences=1, truncation=True)

    full_text = result[0]['generated_text']
    translation = full_text.split(f"{target_lang}:")[-1].strip()
    return translation

fr_translation = translate_with_gpt2(text, "French")
de_translation = translate_with_gpt2(text, "German")

print("--- Original English ---")
print(text)

print("\n--- French Translation (GPT-2) ---")
print(fr_translation)

print("\n--- German Translation (GPT-2) ---")
print(de_translation)
```

```python
# Question 8: Implement a simple attention-based encoder-decoder model for English-to-Spanish translation using Tensorflow or PyTorch.

import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Attention, Concatenate
from tensorflow.keras.models import Model

vocab_size = 5000
embedding_dim = 256
latent_dim = 512
max_len = 20

encoder_inputs = Input(shape=(max_len,), name="Encoder_Input")
encoder_emb = Embedding(vocab_size, embedding_dim)(encoder_inputs)
encoder_outputs, state_h, state_c = LSTM(latent_dim, return_sequences=True, return_state=True)(encoder_emb)

decoder_inputs = Input(shape=(max_len,), name="Decoder_Input")
decoder_emb_layer = Embedding(vocab_size, embedding_dim)
decoder_emb = decoder_emb_layer(decoder_inputs)

decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_emb, initial_state=[state_h, state_c])

attention_layer = Attention()
context_vector = attention_layer([decoder_outputs, encoder_outputs])

decoder_combined_context = Concatenate(axis=-1)([decoder_outputs, context_vector])

outputs = Dense(vocab_size, activation="softmax")(decoder_combined_context)

model = Model([encoder_inputs, decoder_inputs], outputs)
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy")

model.summary()
```
